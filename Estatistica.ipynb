```python
import pandas as pd
from pandas import concat
```

```python
df = pd.read_csv('dados/ESPORTISTAS_MOD7.csv', delimiter = ';')

# EXPLORAÇÃO ESTRUTURAL DOS DADOS
print(df.head(5), '\n')

print(f'Quantidade Linhas: {df.shape[0]}\nQuantidade de Colunas: {df.shape[1]}\n')
print(df.info()) # ---> QTD de registros nulos, não nulos e tipagem de cada coluna
```

## ANÁLISE DOS DADOS

```python
# AGRUPANDO A QUANTIDADE DE ATLETAS DE CADA ESPORTE E DESCOBRINDO SUA REPRESENTATIVIDADE
atletas_por_esporte1 = df.groupby('Esporte')['Atleta'].count().reset_index().sort_values(by = 'Atleta', ascending = False)
total_atletas = atletas_por_esporte1['Atleta'].sum() # --> Retorna o número total de atletas

atletas_por_esporte1['Proporção[%]'] = (atletas_por_esporte1['Atleta'] / total_atletas) * 100 # ---> Adicionando uma nova coluna com a representatividade

print(atletas_por_esporte1, '\n')

# OUTRA FORMA DE CALCULAR O PERCENTUAL - NÃO CONSEGUI ENCAIXAR NO CÓDIGO ACIMA
representatividade = df['Esporte'].value_counts(normalize = True).reset_index()
print(representatividade, '\n')
```

## 2) MEDIANA

```python
# MEDIANA DOS ATLETAS
print(f'Média: R$ {df['Ganhos_Totais'].mean():.0f} milhões/ano')
print(f'A mediana dos atletas é: R$ {df['Ganhos_Totais'].median():.0f} milhões/ano\n')

mediana_ganhos = df.groupby('Esporte')['Ganhos_Totais'].median().reset_index().sort_values(by = 'Ganhos_Totais', ascending = False)
mediana_ganhos.rename(columns = {'Ganhos_Totais': 'Mediana Ganhos'}, inplace = True)

print(f'MEDIANA POR ESPORTE\n{mediana_ganhos}\n\nNOVA TABELA')

# ADICIONANDO A COLUNA DE MEDIANA A TABELA
# atletas_por_esporte1 = atletas_por_esporte1.merge(mediana_ganhos, on = 'Esporte')

atletas_por_esporte1
```

## 3) MODA

```python
# TIRANDO A MODA DE ESPORTE
moda = df['Esporte'].mode()
print(f'A moda é: {moda[0]}')
```

## 4) DESVIO PADRÃO

- Alguns valores é normal dar erro (NaN) por conta que o desvio padrão precisa de múltiplos valores para ser calculado

```python
print(f'O desvio padrão dos salários é de: R$ {df['Ganhos_Totais'].std():.0f} milhões/ano')

std_ganhos = df.groupby('Esporte')['Ganhos_Totais'].std().reset_index().sort_values(by = 'Ganhos_Totais', ascending = False)
std_ganhos.rename(columns = {'Ganhos_Totais': 'Desvio Padrão'}, inplace = True)
std_ganhos['Desvio Padrão'] = std_ganhos['Desvio Padrão'].round(2) # ---> ARREDONDANDO OS VALORES DO DESVIO PADRÃO
print(f'DESVIO PADRÃO POR ESPORTE\n{std_ganhos}\n\nNOVA TABELA')

# atletas_por_esporte1 = atletas_por_esporte1.merge(std_ganhos, on = 'Esporte')

# TRATANDO OS VALORES QUE DERAM ERRO
atletas_por_esporte1['Desvio Padrão'] = atletas_por_esporte1['Desvio Padrão'].fillna(0)

atletas_por_esporte1
```
